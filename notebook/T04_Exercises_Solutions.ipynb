**2**. Download the attention dataset from https://bit.ly/2B1iGZB, treating column 1 as the index, and answer the following:

1. How many rows and columns are there?
2. Drop the `solutions` column
3. Sort by `score` in descending order
4. Who is the subject with the highest score in divided attention?
5. Show the mean and standard deviations of the scores by subject and attention.
6. Find the difference between the scores for each subject when solution = 1 and when solution = 2.

```python
attention = pd.read_csv('https://bit.ly/2B1iGZB', index_col=0)
```

```python
attention
```

```python
attention.head(3)
```

```python
attention.shape
```

```python
attention = attention.drop('solutions', axis=1)
```

```python
attention.head(3)
```

```python
attention[attention.attention=='divided'].nlargest(1, 'score')
```

```python
attention.groupby(['subject', 'attention'])[['score']].agg(['mean', 'std'])
```

```python
attention = attention[attention.solutions.isin([1,2])]
```

```python
attention.shape
```

```python
attention= attention.drop('attention', axis=1)
```

```python
attention.head(3)
```

```python
df = attention.pivot(index='subject', columns='solutions', values='score')
```
